```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi, 0.1)

plt.subplot(1,3,1)
plt.plot( x, np.sin(x) )

plt.subplot(1,3,3)
plt.plot( x, np.cos(x) )

plt.subplot(1,3,2)
plt.plot( x, np.tan(x) )
```
